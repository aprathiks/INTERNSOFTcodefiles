```python
import math

def is_prime(num):
    if num % 2 == 0 and num > 2:
        return False
    for i in range(3,int(math.sqrt(num))+1,2):
        if num % i == 0:
            return False
    return True
```

```python
def fibon(n):
    a=1
    b=1
    output=[]
    
    for i in range(n):
        output.append(a)
        a,b=b,a+b
    return output
```
